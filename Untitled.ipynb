```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")
%config IPCompleter.greedy=True
```

```python
df = pd.read_csv('monthly-data-relating-to-hotels-.csv')
```

```python
df['Room_nights'] = pd.to_numeric(df['Room_nights'],errors='coerce')
df['Month'] = pd.to_datetime((df['Month']), errors='coerce')
```

```python
df.info()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.info()
```

```python
df = df.set_index('Month')
df.index
```

```python
df.tail()
```

```python
nights = df['Room_nights']
```

```python
nights.plot(figsize=(15, 6))
from pylab import rcParams
rcParams['figure.figsize'] = 18, 8
plt.show()
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 18, 18

decomposition = sm.tsa.seasonal_decompose(nights, model='additive')
fig = decomposition.plot()
plt.show()
```

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(nights,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

ARIMA(0, 1, 1)x(1, 1, 1, 12)12 - AIC:3496.0972847440526

```python
mod = sm.tsa.statespace.SARIMAX(nights,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()
print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(16, 12))
plt.show()
```

```python
pred = results.get_prediction(start=pd.to_datetime('1992-01'), dynamic=False)
pred_ci = pred.conf_int()
ax = nights['1988-01':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Nights')
plt.legend()
plt.show()
```

```python
y_forecasted = pred.predicted_mean
y_truth = nights['1992-01':]
mse = ((y_forecasted - y_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))
```

```python
print('The Root Mean Squared Error of our forecasts is {}'.format(round(np.sqrt(mse), 2)))
```

```python
from fbprophet import Prophet
```

```python
test_nights = nights['1992':]
```

```python
new_nights = nights[:'1991']
new_nights.tail()
```

```python

prof_nights = pd.DataFrame()
prof_nights['ds'] = new_nights.index
prof_nights['y'] = new_nights.values
```

```python
prof_nights.head()
```

```python
nights_model = Prophet(interval_width=0.95)
nights_model.fit(prof_nights)
```

```python
forecast = nights_model.make_future_dataframe(periods=42, freq='MS')
forecast = nights_model.predict(forecast)
```

```python
plt.figure()
nights_model.plot(forecast, xlabel='Date', ylabel='Nights')
plt.title('Nights Prophet forecast')
```

```python
nights_model.plot_components(forecast);
```
